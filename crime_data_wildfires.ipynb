Wildfire Event Catgorization

```python
#import libraries and dependencies
import pandas as pd
import numpy as np
import os

```

Franklin Fire - 12.9.2024-12.18.2024
4,037 acres burned

```python
#read in the data
wildfire_data_2024 = pd.read_csv('resources/cleaned_data/crime_2024.csv')

#display the data
wildfire_data_2024
```

```python
#convert the DATE OCC column to datetime
wildfire_data_2024['DATE OCC'] = pd.to_datetime(wildfire_data_2024['DATE OCC'])

#define the date range for the wildfires
start_date = '2024-12-09'
end_date = '2024-12-25'

#create a new column to flag the wildfires
wildfire_data_2024['Wildfire Event'] = wildfire_data_2024['DATE OCC'].apply(lambda date: '1' if start_date <= date.strftime('%Y-%m-%d') <= end_date else '0')

#write the data into a new csv file
wildfire_data_2024.to_csv('resources/cleaned_data/wildire_data/crime_2024_wildfire.csv', index=False)
```
