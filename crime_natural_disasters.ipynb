```python
#import dependencies
import pandas as pd
from pathlib import Path
import os
import numpy as np

#read in the data
file_path = Path('Resources/cleaned_data/crime_2022.csv')
crime_data_2022 = pd.read_csv(file_path)

#check the data
crime_data_2022.head()
```

```python
#find all incidents that occured in september

crime_data_september = crime_data_2022.loc[crime_data_2022['DATE OCC'].dt.month == 9]

#check the data
crime_data_september.head()
```

```python
#write the data to a csv file
crime_data_september.to_csv('Resources/cleaned_data/crime_september_2022.csv', index=False)
```
